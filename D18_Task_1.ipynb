```python
def login():
  f=open("data.txt","r")
  print("LOGIN")
  uname=str(input("Enter Emailid:"))
  pwd=str(input("Enter Password:"))
  l=f.readlines()
  l2=[]
  for i in l:
   x=i.split()
   l2+=x   
  if uname in l2 and pwd in l2:
      print(" welcome",uname[:uname.index('@')]) 
      exit()
  if uname not in l2 or pwd not in l2:
      print("Oops! User doesn't exist or the password is incorrect")
      print("1. REGISTER")
      print("2. FORGOT PASSWORD!")
      print("3. Retry")
      print("please input a number ")
      usrip=int(input())
      if usrip==1:
        r=registration()
        if r==1:
          print("registration successful now login")
          login()
      if usrip == 2:
        forgotpassword()
      if usrip==3:
        login()
      
      
    
    
    
def forgotpassword():
      
      print("Forgot Password")
      print("Please Enter Your registerd mail id:")
      email=str(input())
      f=open("data.txt","r")
      l=f.readlines()
      l2=[]
      for i in l:
        x=i.split()
        l2+=x
      if(email in l2):
          print("your password is :",l2[l2.index(email)+1])
          f.close()
          login()
          
      else:
         print("user doesn't exist would you like to register?(y/n)")
         r=input()
         if r =='y':
           y=registration()
           if y==1:
            login()
         else:
           forgotpassword()






def registration():
  print("Registration")
  while(True):
    uname=str(input("Enter Mail Id:"))
    pwd=str(input("Enter Password"))
    r=check(uname)
    p=pcheck(pwd)
    if r==1 or p==1:
      print('invalid email or Password')
    else:
      f=open("data.txt","a")
      f.write (uname +' ')
      f.write(pwd+'\n')
      f.close()
      return 1
      break
  

def check(email):
  
  if '@' in email and '.' in email:
    if email[0]!='@':
      if email.index('@')+1!=email.index('.'):
        if '#'or'$'or'!'or'%'or'^'or'*'or'&'or'*'or'_'or'-'or'+'or'='or'/'or '{'or '}'not in email[0] and email[0].isdigit()!=True:
           return 0
        else:
          return 1
      else:
        return 1
    else:
      return 1      

  else:
    return 1

def pcheck(password):
  ct=0
  spc=['@','!','^','#','%','$','*','_','{','}','[',']','(',')','?','<','>','/','|','~']
  for i in password:
    if i.isdigit()==True :
      ct+=1
    if i.islower()==True:
      ct+=1
    if i.isupper()==True:
      ct+=1
    if i in spc:
      ct+=1
    
  if ct>=4 and len(password)>5 and len(password)<16:
    return 0
  else:
    return 1


print("TASK-1 Registration and Login system using Python, file handling")
print("1.LOGIN ")
print("2.REGISTER / SIGNUP ")
print("please input a number ")
f=open("data.txt","a")
f.close()
inp=int(input())
if inp==1:
  login()
if inp==2:
  r=registration()
  if r==1:
    print("registration successful now login")
    login()


```
